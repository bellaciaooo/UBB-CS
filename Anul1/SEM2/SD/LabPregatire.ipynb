```python
#Exercitiul 2
#solutia generala
reset()
t = var('t')
x = function('x')(t)
y = function('y')(t)

deq1 = diff(x,t) == 2*x - y
deq2 = diff(y,t) == 3*x - 2*y
syst = [deq1,deq2]
desolve_system(syst,[x,y])
```

```python
#b) sol ce satisface x0=2,y0=4 si reprezentati grafic in aceasi fereastra pe [-3,3]
desolve_system(syst,[x,y],[0,2,4])
```

```python
sol = desolve_system(syst,[x,y],[0,2,4])
sol_x(t) = sol[0].rhs()
sol_y(t) = sol[1].rhs()

g1 = plot(sol_x(t,2,4),t,-3,3,color = 'red')
g2 = plot(sol_y(t,2,4),t,-3,3,color = 'blue')
show(g1+g2)
```
