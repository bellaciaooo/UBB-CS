```python
#Exercitiul 1
#A
x = var('x')
y = function('y')(x)
eqd = diff(y,x)==2*x*(1+y^2)
ec = desolve(eqd,y)
show(solve(ec,y))
```

```python
#B
x = var('x')
y = function('y')(x)
eqd = (x^2-1)*diff(y,x)+2*x*y^2==0
ec = desolve(eqd,y)
show(solve(ec,y))
```

```python
#C
x = var('x')
y = function('y')(x)
eqd = 2*x^2*diff(y,x)==x^2+y^2
ec = desolve(eqd,y)
show(solve(ec,y))
```

```python
#D
x = var('x')
y = function('y')(x)
eqd = diff(y,x)==-(x+y)/y
ec = desolve(eqd,y)
show(ec)
```

```python
#E
x = var('x')
y = function('y')(x)
eqd = diff(y,x,2)+y==sin(x)+cos(x)
ec = desolve(eqd,y)
show(ec)
```

```python
#F
x = var('x')
y = function('y')(x)
eqd = diff(y,x,2)-y==exp(2*x)
ec = desolve(eqd,y)
show(ec)
```

```python
#G
x = var('x')
y = function('y')(x)
eqd = diff(y,x,2)-diff(y,x)==1/(1+exp(x))
ec = desolve(eqd,y)
show(ec)
```

```python
#Exercitiul 2
#A
x = var('x')
y = function('y')(x)
eqd = diff(y,x)==1+y^2
ec = desolve(eqd,y,ics=[0,1])
rez = solve(ec,y)
rez
```

```python
sol=rez[0].rhs()
```

```python
plot(sol,x,-2,4,detect_poles='True',ymin=-10,ymax=10)
```

```python
#B
x = var('x')
y = function('y')(x)
eqd = diff(y,x)==1/(1-x^2)*y+1+x
ec = desolve(eqd,y,ics=[2,0])
ec
```

```python
plot(ec,0,10,detect_poles='True')
```
