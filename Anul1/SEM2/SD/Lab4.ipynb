```python
#Exercitiul 1: Modelul dezintegrarii radioactive
#(a) Sa se determine solutia problemei Cauchy
t, k = var('t,k')
x = function('x')(t)
eqd = diff(x,t) == -k*x
sol(t,k,x0) = desolve(eqd,[x,t],[0,x0])
show(sol)
```

```python
#(b) Reprezentati grafic cateva solutii
plot([sol(t,0.1,10),sol(t,0.2,10)],t,0,10)
```

```python
#(c) Stiind ca timpul de injumatatire al izotopului radioactiv C14 este t1/2 = 5730 ani sa se determine constanta k.
# Timpul de injumatatire reprezinta timpul in care o cantitate de C14 se reduce la jumatate in urma dezintegrarii radioactive.
T12 = var('T12')
ec = sol(T12,k,x0) == x0/2
ans = solve(ec,k)
show(ans)
```

```python
#(d)  Folosind constanta k determinata, sa se estimeze vechimea unor fosile stiind ca in momentul descoperirii raportul 
# C14/C12 a ajuns la 20% .
k1(T12) = ans[0].rhs()
kC14 = k1(5730)
#numerical_approx(kC14) -verificam valoarea lui kC14
T = var('T')
ec = sol(T,kC14,1) == 0.2
show(solve(ec,T))
```

```python
ans = solve(ec,T)
numerical_approx(ans[0].rhs())
```
