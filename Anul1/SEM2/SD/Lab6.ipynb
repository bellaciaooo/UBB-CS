```python
#Exercitiul 1: Sa se determine solutiile de echilibru pentru fiecare din ecuatiile autonome date.
#a)
reset()
t = var('t')
x = function('x')(t)

s = var('s')
f = function('f')(s)
f(s) = s^2 - 2*s

eqp = solve(f(s) == 0, s)
show(eqp)
```

```python
x1 = eqp[0].rhs()
x2 = eqp[1].rhs()
```

```python
diff(f,s)(x1)
```

```python
diff(f,s)(x2)
```

```python
sf = plot_slope_field(f(s), (t,-5,5), (s,-3,3), headaxislength = 3, headlength = 4, color = 'red')
sf
```

```python
deq = diff(x,t) == f(x)
deq
```

```python
desolve_rk4(deq, x, [0,0.5], step = 0.5, end_points = [-5,5], output='list')
```

```python
desolve_rk4(deq, x, [0,0.5], step = 0.1, end_points = [-5,5], output='plot')
```

```python
desolve_rk4(deq, x, [0,0.5], step = 0.1, end_points = [-5,5], output='slope_field', color = 'orange')
```
